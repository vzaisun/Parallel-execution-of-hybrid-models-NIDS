```python
#One-hot encoding
def one_hot(df, cols):
    """
    @param df pandas DataFrame
    @param cols a list of columns to encode
    @return a DataFrame with one-hot encoding
    """
    for each in cols:
        dummies = pd.get_dummies(df[each], prefix=each, drop_first=False)
        df = pd.concat([df, dummies], axis=1)
        df = df.drop(each, 1)
    return df
```

```python
#Function to min-max normalize
def normalize(df, cols):
    """
    @param df pandas DataFrame
    @param cols a list of columns to encode
    @return a DataFrame with normalized specified features
    """
    result = df.copy() # do not touch the original df
    for feature_name in cols:
        max_value = df[feature_name].max()
        min_value = df[feature_name].min()
        if max_value > min_value:
            result[feature_name] = (df[feature_name] - min_value) / (max_value - min_value)
    return result
```

```python
from keras.layers import Input, Concatenate, Dense, Activation, Dropout, Conv1D, MaxPooling1D, BatchNormalization, Bidirectional, LSTM
from keras.models import Model
from keras.utils import to_categorical

# Define input shape for both branches
input_shape = (122, 1)

# Create input layers for each branch
input_cnn_lstm = Input(shape=input_shape, name='input_cnn_lstm')
input_cnn_bilstm = Input(shape=input_shape, name='input_cnn_bilstm')

# CNN-LSTM branch
cnn_lstm = Conv1D(64, kernel_size=122, padding="same", activation="relu")(input_cnn_lstm)
cnn_lstm = MaxPooling1D(pool_size=5)(cnn_lstm)
cnn_lstm = BatchNormalization()(cnn_lstm)
cnn_lstm = LSTM(64, return_sequences=False)(cnn_lstm)
cnn_lstm = Dropout(0.5)(cnn_lstm)

# CNN-BiLSTM branch
cnn_bilstm = Conv1D(64, kernel_size=122, padding="same", activation="relu")(input_cnn_bilstm)
cnn_bilstm = MaxPooling1D(pool_size=5)(cnn_bilstm)
cnn_bilstm = BatchNormalization()(cnn_bilstm)
cnn_bilstm = Bidirectional(LSTM(64, return_sequences=False))(cnn_bilstm)
cnn_bilstm = Dropout(0.5)(cnn_bilstm)

# Concatenate the outputs of both branches
merged = Concatenate(axis=-1)([cnn_lstm, cnn_bilstm])

# Fully connected layers
output = Dense(5, activation="softmax")(merged)

# Create the hybrid model
hybrid_model = Model(inputs=[input_cnn_lstm, input_cnn_bilstm], outputs=output)

# Compile the hybrid model with 'sparse_categorical_crossentropy' loss
hybrid_model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

```
